```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
data = pd.read_csv(r"C:\Users\GASTO\Downloads\Breast_cancer_data.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
from  sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(data)
scaled_data=scaler.transform(data)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca.fit(scaled_data)
```

```python
x_pca=pca.transform(scaled_data)
print(scaled_data.shape)
print(x_pca.shape)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=data['diagnosis'],cmap='plasma')
plt.xlabel('First principal component')
plt.ylabel('Second principal component')
plt.show()
```

```python
# interpretenting the components
pca.components_
```

they give the weights

```python
feature_names=data.columns
feature_names
```

```python
# heat map
df_comp= pd.DataFrame(pca.components_,columns=feature_names)
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma')
```

represent the correclation between the features and the principal component
